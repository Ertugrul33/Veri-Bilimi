```python
liste = range(100)
yeniListe = []

for i in liste:
    if i % 5 == 0:
        if i % 10 == 0:
            yeniListe.append(i**2)
yeniListe
```

```python
yeniListe = [i**2 for i in liste if i % 5 == 0 if i % 10 == 0]
yeniListe
```
